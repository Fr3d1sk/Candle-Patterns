In this example I show how someone could use this script with different settings.
For an advanced usage it is advisable to check the files.

Starting with all colors of candles and with just 1% difference:

```python
import visualization
from Candlestick_patterns import ShootingStar, Hammer, HangingMan, Doji, GravestoneDoji

timeframes = [
    "1h",
    "4h",
    "1d"
]

patterns = [
    ShootingStar(),
    Hammer(),
    Doji(),
    GravestoneDoji(),
    HangingMan()
]
visualization.check_data(timeframes, patterns, image_name="all_candles")
```

Starting with red colored candles and with just 1% difference:

```python
visualization.check_data(timeframes,patterns, candle_type="red", image_name="red_candles")
```

Starting with green colored candles and with just 1% difference:

```python
visualization.check_data(timeframes,patterns, candle_type="green", image_name="green_candles")
```
